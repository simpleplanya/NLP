```python
from transformers import BertTokenizer
```

```python
PRETRAINED_MODEL_NAME = "bert-base-multilingual-cased"  
# 載入BERT預訓練模型
tokenizer = BertTokenizer.from_pretrained(PRETRAINED_MODEL_NAME)
```

```python
vocab = tokenizer.vocab
print("字典大小：", len(vocab))
```

```python
print('字典內容')
vocab
```

```python
text = "Rock"
tokens = tokenizer.tokenize(text)
ids = tokenizer.convert_tokens_to_ids(tokens)
print(text)
print(tokens, '...')
print(ids, '...')
```

```python
#測試是否OOV
text = "OOC"
tokens = tokenizer.tokenize(text)
ids = tokenizer.convert_tokens_to_ids(tokens)
print(text)
print(tokens, '...')
print(ids, '...')
```

```python
#測試是否OOV
text = "藬"
tokens = tokenizer.tokenize(text)
ids = tokenizer.convert_tokens_to_ids(tokens)

print(text)
print(tokens, '...')
print(ids, '...')
```

```python
# 新增token到字典
tokenizer.add_tokens(['OOC'])
#回傳的數字代表被加入的token數目(因為list輸入不只一個token)
```
